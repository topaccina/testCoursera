```python
import json
import pandas as pd
import re
```

```python
f = open('cfp-export-2023-01-31.json')
data = json.load(f)
```

```python
title=[]
milestones=[]
milestonesStatus=[]
```

```python
for cfp in data['checklists']:
    # print(cfp['name'])
    if re.match(r"Milestones CFP", cfp['name']):
        title.append(cfp['name'])
        cfpStep=[]
        cfpStepStatus=[]
        for step in cfp['checkItems']:
            # print(step['name'])
            cfpStep.append(step['name'])
            cfpStepStatus.append(step['state'])
        milestones.append(cfpStep)
        milestonesStatus.append(cfpStepStatus)
```

```python
df=pd.DataFrame(data={'title':title,'milestones':milestones,'milestones_status':milestonesStatus})
```
